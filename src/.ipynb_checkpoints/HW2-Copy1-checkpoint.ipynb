```python
# Imports
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import folium
from folium.features import GeoJsonPopup, GeoJsonTooltip
```

```python
# Leer CSV 
hp_df = pd.read_csv("hospitals_info.csv") 

# Crear geometría (asegúrate de usar lon primero, luego lat) 
hp_df["geometry"] = hp_df.apply(lambda row: Point(row["ESTE"], row["NORTE"]), axis=1)
```

```python
# --- 1. Convertir hospitales a GeoDataFrame ---
hp_gdf = gpd.GeoDataFrame(
    hp_df,
    geometry=hp_df.apply(lambda row: Point(row["NORTE"], row["ESTE"]), axis=1),
    crs="EPSG:4326"

)

hp_gdf
```

```python
# --- 2. Asegurarse que pop_gdf tenga CRS ---
print(pop_gdf.crs)  # revisar, probablemente EPSG:4326
```

```python
# --- 3. Reproyectar a CRS métrico (UTM zona 18S) ---
pop_utm = pop_gdf.to_crs(epsg=32718)
hp_utm = hp_gdf.to_crs(epsg=32718)
```

```python
# --- 4. Crear buffer de 10 km alrededor de cada centro poblado ---
pop_utm["buffer_10km"] = pop_utm.geometry.buffer(10000)  # 10,000 m = 10 km
```

```python
# --- 5. Contar hospitales dentro de cada buffer ---
buffers_gdf = pop_utm.set_geometry("buffer_10km")  # geometría activa = buffer
join = gpd.sjoin(hp_utm, buffers_gdf, predicate="within")

# Contar hospitales por cada centro poblado
hospital_counts = join.groupby(join.index_right).size()

# --- 6. Agregar la cuenta al GeoDataFrame de población ---
pop_utm["hospitals_within_10km"] = pop_utm.index.map(hospital_counts).fillna(0).astype(int)
```

```python
# --- 7. Reproyectar de nuevo a EPSG:4326 para mapas interactivos (Folium) ---
pop_gdf_final = pop_utm.to_crs(epsg=4326)
```

```python
pop_utm
```

```python
# --- 1. Filtrar solo Lima y Loreto ---
regiones = ["LIMA", "LORETO"]
pop_sub = pop_utm[pop_utm["Departamento"].isin(regiones)].copy()

# --- 2. Repetir el spatial join solo para estos departamentos ---
buffers_sub = pop_sub.set_geometry("buffer_10km")

# Spatial join hospitales dentro de buffers
join_sub = gpd.sjoin(hp_utm, buffers_sub, predicate="within")

# Contar hospitales por cada centro poblado
hospital_counts_sub = join_sub.groupby(join_sub.index_right).size()

# --- 3. Agregar la cuenta ---
pop_sub["hospitals_within_10km"] = pop_sub.index.map(hospital_counts_sub).fillna(0).astype(int)

# --- 4. Identificar aislamiento y concentración por región ---
results = {}

for reg in regiones:
    df_reg = pop_sub[pop_sub["Departamento"] == reg].copy()
    
    # Centro poblado con menos hospitales
    isolation = df_reg.loc[df_reg["hospitals_within_10km"].idxmin()]
    
    # Centro poblado con más hospitales
    concentration = df_reg.loc[df_reg["hospitals_within_10km"].idxmax()]
    
    results[reg] = {
        "isolation": isolation,
        "concentration": concentration,
        "summary": df_reg[["NOM_POBLAD", "hospitals_within_10km"]].sort_values("hospitals_within_10km", ascending=False)
    }

# --- 5. Mostrar resultados ---
for reg, info in results.items():
    print(f"\n🔹 Región: {reg}")
    print(f"   📍 Mayor concentración: {info['concentration']['NOM_POBLAD']} ({info['concentration']['hospitals_within_10km']} hospitales en 10km)")
    print(f"   📍 Mayor aislamiento: {info['isolation']['NOM_POBLAD']} ({info['isolation']['hospitals_within_10km']} hospitales en 10km)")
```

```python
# ---------- Task 1: Choropleth nacional (distrito) + MarkerCluster ----------
# 1) Contar hospitales por distrito via spatial join
# Aseguramos que districts tenga la columna id como string
districts_gdf[district_id_col] = districts_gdf[district_id_col].astype(str)

# Spatial join: asignar cada hospital a un distrito (requiere que ambos estén en EPSG:4326)
join = gpd.sjoin(hp_gdf, districts_gdf[[district_id_col,"geometry"]], how="left", predicate="within")
counts = join.groupby(district_id_col).size().reset_index(name="num_hospitals")

# Merge conteos con shapefile (llenar 0)
districts_count = districts_gdf.merge(counts, on=district_id_col, how="left").fillna({"num_hospitals":0})
districts_count["num_hospitals"] = districts_count["num_hospitals"].astype(int)

# Convertir a geojson para folium
districts_geojson = districts_count.to_crs(epsg=4326).to_json()

# 2) Crear mapa folium y choropleth
m_choro = folium.Map(location=[-9.3, -75.0], zoom_start=5, tiles="cartodbpositron")  # centro aproximado del Perú

folium.Choropleth(
    geo_data=districts_geojson,
    data=districts_count,
    columns=[district_id_col, "num_hospitals"],
    key_on=f"feature.properties.{district_id_col}",
    fill_color="YlOrRd",
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name="Número de hospitales por distrito",
    nan_fill_color="white",
).add_to(m_choro)

# 3) MarkerCluster con todos los hospitales
marker_cluster = MarkerCluster(name="Hospitales").add_to(m_choro)
for _, h in hp_gdf.iterrows():
    name = h.get(hosp_name_col, "Hospital")
    popup = folium.Popup(f"{name}", parse_html=True)
    folium.CircleMarker(
        location=[h.geometry.y, h.geometry.x],
        radius=1,
        popup=popup,
        fill=True,
        fill_opacity=0.7
    ).add_to(marker_cluster)

# Guardar mapa
m_choro.save("national_hospitals_choropleth.html")
print("Mapa choropleth guardado: national_hospitals_choropleth.html")
```

```python
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# ---------- Configuración ----------
districts_shp_path = "DISTRITOS.shp"   # Shapefile de distritos
district_id_col   = "IDDIST"           # ID distrito en el shapefile
hp_x_col          = "NORTE"            # Longitud en hospitales
hp_y_col          = "ESTE"             # Latitud en hospitales
pop_name_col      = "NOM_POBLAD"       # Nombre del centro poblado
hosp_name_col     = "Nombre del establecimiento"
department_col    = "Departamento"

# ---------- Cargar datos ----------
districts_gdf = gpd.read_file(districts_shp_path).to_crs(epsg=4326)

# Aseguramos que el ID de distrito sea string
districts_gdf[district_id_col] = districts_gdf[district_id_col].astype(str)

# ---------- Spatial Join: hospitales dentro de distritos ----------
join = gpd.sjoin(
    hp_gdf, 
    districts_gdf[[district_id_col, "geometry"]], 
    how="left", 
    predicate="within"
)

# Contar hospitales por distrito
counts = join.groupby(district_id_col).size().reset_index(name="num_hospitals")

# Unir conteos al shapefile
districts_count = districts_gdf.merge(counts, on=district_id_col, how="left").fillna({"num_hospitals": 0})
districts_count["num_hospitals"] = districts_count["num_hospitals"].astype(int)

# Convertir a GeoJSON para Folium
districts_geojson = districts_count.to_json()

# ---------- Crear mapa base ----------
m_choro = folium.Map(location=[-9.3, -75.0], zoom_start=5, tiles="cartodbpositron")

# ---------- Choropleth ----------
folium.Choropleth(
    geo_data=districts_geojson,
    data=districts_count,
    columns=[district_id_col, "num_hospitals"],
    key_on=f"feature.properties.{district_id_col}",
    fill_color="YlOrRd",
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name="Número de hospitales por distrito",
    nan_fill_color="white",
).add_to(m_choro)

# ---------- MarkerCluster con hospitales ----------
marker_cluster = MarkerCluster(name="Hospitales").add_to(m_choro)

for _, h in hp_gdf.iterrows():
    name = h.get(hosp_name_col, "Hospital")
    distrito = h.get("Distrito", "Desconocido")
    popup_html = f"<b>{name}</b><br>Distrito: {distrito}"
    folium.Marker(
        location=[h.geometry.y, h.geometry.x],
        popup=popup_html,
        icon=folium.Icon(color="red", icon="plus-sign")
    ).add_to(marker_cluster)

# ---------- Guardar ----------
m_choro.save("national_hospitals_choropleth.html")
print("✅ Mapa choropleth guardado: national_hospitals_choropleth.html")
```

```python
# -------------------------
# Proximity visualization: Lima & Loreto (Folium)
# -------------------------
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from shapely.geometry import Point, mapping

# ---------- Helper: elegir EPSG UTM según lon/lat ----------
def lonlat_to_utm_epsg(lon, lat):
    """Devuelve EPSG UTM (hemisferio sur -> 327##) para una lon/lat dada."""
    zone = int((lon + 180) // 6) + 1
    return 32700 + zone  # EPSG 327## (sur)

# ---------- Función principal ----------
def make_proximity_map(region_name,
                       pop_gdf,
                       hp_gdf,
                       buffer_m=10000,
                       map_filename=None,
                       pop_name_col="NOM_POBLAD",
                       hosp_name_col="Nombre del establecimiento",
                       departamento_col="Departamento"):
    """
    Crea y guarda un mapa folium para la región (LIMA o LORETO).
    Devuelve dict con: map object, pop_reg (WGS84), buffers_wgs84, hospitals_in_buffers (WGS84), isolation, concentration.
    """
    region = region_name.upper()
    # 0) Asegurar CRS WGS84 en inputs
    if pop_gdf.crs is None:
        pop_gdf = pop_gdf.set_crs(epsg=4326)
    else:
        pop_gdf = pop_gdf.to_crs(epsg=4326)
    if hp_gdf.crs is None:
        hp_gdf = hp_gdf.set_crs(epsg=4326)
    else:
        hp_gdf = hp_gdf.to_crs(epsg=4326)

    # 1) Filtrar centros poblados de la región
    pop_reg = pop_gdf[pop_gdf[departamento_col].str.upper() == region].copy()
    if pop_reg.empty:
        print(f"No se encontraron centros poblados para la región: {region_name}")
        return None

    # 2) Elegir UTM basado en centroid del conjunto (mejor para buffers métricos)
    lon_center = pop_reg.geometry.x.mean()
    lat_center = pop_reg.geometry.y.mean()
    utm_epsg = lonlat_to_utm_epsg(lon_center, lat_center)

    # 3) Reproyectar a UTM métrico
    pop_reg_utm = pop_reg.to_crs(epsg=utm_epsg)
    hp_reg_utm = hp_gdf.to_crs(epsg=utm_epsg)

    # 4) Crear buffer de 10 km (si ya existe columna buffer_10km en pop_reg_utm, la reutiliza)
    if "buffer_10km" not in pop_reg_utm.columns:
        pop_reg_utm["buffer_10km"] = pop_reg_utm.geometry.buffer(buffer_m)

    # 5) Preparar buffers gdf y hacer spatial join (intersects para incluir bordes)
    buffers = gpd.GeoDataFrame(
        pop_reg_utm[[pop_name_col]].copy(),
        geometry=pop_reg_utm["buffer_10km"],
        crs=pop_reg_utm.crs
    ).reset_index(drop=True)
    buffers = buffers.reset_index().rename(columns={"index":"buffer_index"})  # índice para referencia

    # join hospitales ↔ buffers
    joined = gpd.sjoin(hp_reg_utm, buffers[["buffer_index","geometry"]], how="inner", predicate="intersects")

    # contar hospitales por buffer_index
    if joined.shape[0] > 0:
        counts = joined.groupby("buffer_index").size().reset_index(name="n_hosp")
    else:
        counts = pd.DataFrame(columns=["buffer_index","n_hosp"])

    buffers = buffers.merge(counts, on="buffer_index", how="left").fillna({"n_hosp":0})
    buffers["n_hosp"] = buffers["n_hosp"].astype(int)

    # 6) Unir conteo de vuelta a pop_reg_utm (por orden)
    # Aseguramos índice coincidente con bufs: tomamos el mismo orden que buffers (buffer_index)
    pop_reg_utm = pop_reg_utm.reset_index(drop=True)
    pop_reg_utm = pop_reg_utm.merge(buffers[["buffer_index","n_hosp"]], left_index=True, right_on="buffer_index", how="left")
    pop_reg_utm["n_hosp"] = pop_reg_utm["n_hosp"].fillna(0).astype(int)

    # 7) Identificar aislamiento y concentración
    idx_min = pop_reg_utm["n_hosp"].idxmin()
    idx_max = pop_reg_utm["n_hosp"].idxmax()
    isolation = pop_reg_utm.loc[idx_min]
    concentration = pop_reg_utm.loc[idx_max]

    # 8) Extraer hospitales dentro de cada buffer (en UTM)
    hosp_in_isolation = hp_reg_utm[hp_reg_utm.geometry.within(isolation["buffer_10km"])]
    hosp_in_concentration = hp_reg_utm[hp_reg_utm.geometry.within(concentration["buffer_10km"])]

    # 9) Reproyectar todo a WGS84 para Folium
    buffers_wgs84 = buffers.set_geometry("geometry").to_crs(epsg=4326)
    pop_reg_wgs84 = pop_reg_utm.to_crs(epsg=4326)
    hp_in_iso_wgs84 = hosp_in_isolation.to_crs(epsg=4326)
    hp_in_con_wgs84  = hosp_in_concentration.to_crs(epsg=4326)

    # 10) Preparar mapa folium centrado en la región
    center = [pop_reg_wgs84.geometry.y.mean(), pop_reg_wgs84.geometry.x.mean()]
    m = folium.Map(location=center, zoom_start=7, tiles="CartoDB positron")

    # 11) Añadir marker cluster con todas las hospitals de la región (opcional)
    cluster_all = MarkerCluster(name="Hospitales (región)").add_to(m)
    # Para rendimiento, sólo añadimos hospitales que caen en bbox ampliada de la región
    lon_min, lat_min, lon_max, lat_max = pop_reg_wgs84.total_bounds
    # filtrar hp por bbox primero (WGS84)
    hp_wgs84_region = hp_gdf.to_crs(epsg=4326).cx[lon_min:lon_max, lat_min:lat_max]
    for _, h in hp_wgs84_region.iterrows():
        popup_h = f"<b>{h.get(hosp_name_col, 'Hospital')}</b><br>Distrito: {h.get('Distrito','-')}"
        folium.Marker(location=[h.geometry.y, h.geometry.x],
                      popup=folium.Popup(popup_h, max_width=300),
                      icon=folium.Icon(icon="plus-sign", color="red")
                     ).add_to(cluster_all)

    # 12) Añadir buffer de aislamiento (rojo) y su centroid marker
    # Encontrar geometrías en WGS84 para isolation y concentration
    iso_poly_wgs84 = gpd.GeoSeries([isolation["buffer_10km"]], crs=pop_reg_utm.crs).to_crs(epsg=4326).iloc[0]
    conc_poly_wgs84 = gpd.GeoSeries([concentration["buffer_10km"]], crs=pop_reg_utm.crs).to_crs(epsg=4326).iloc[0]

    # Popup HTMLs (detalle formateado)
    iso_popup_html = f"""
        <div style="font-size:13px">
          <b>Centro:</b> {isolation.get(pop_name_col, 'NA')}<br>
          <b>Departamento:</b> {isolation.get(departamento_col, 'NA')}<br>
          <b>Hospitales en 10 km:</b> {int(isolation['n_hosp'])}
        </div>
    """
    conc_popup_html = f"""
        <div style="font-size:13px">
          <b>Centro:</b> {concentration.get(pop_name_col, 'NA')}<br>
          <b>Departamento:</b> {concentration.get(departamento_col, 'NA')}<br>
          <b>Hospitales en 10 km:</b> {int(concentration['n_hosp'])}
        </div>
    """

    # Añadir polígonos con estilo (rojo / verde)
    folium.GeoJson(data=mapping(iso_poly_wgs84),
                   style_function=lambda feat: {"fillColor":"#ff6666","color":"#cc0000","weight":2,"fillOpacity":0.25}
                  ).add_child(folium.Popup(iso_popup_html, max_width=300)).add_to(m)

    folium.GeoJson(data=mapping(conc_poly_wgs84),
                   style_function=lambda feat: {"fillColor":"#66cc66","color":"#006600","weight":2,"fillOpacity":0.25}
                  ).add_child(folium.Popup(conc_popup_html, max_width=300)).add_to(m)

    # 13) Añadir centroid markers (más visibles)
    iso_centroid = iso_poly_wgs84.centroid
    conc_centroid = conc_poly_wgs84.centroid

    folium.Marker(location=[iso_centroid.y, iso_centroid.x],
                  popup=folium.Popup(iso_popup_html, max_width=300),
                  icon=folium.Icon(color="red", icon="info-sign")
                 ).add_to(m)

    folium.Marker(location=[conc_centroid.y, conc_centroid.x],
                  popup=folium.Popup(conc_popup_html, max_width=300),
                  icon=folium.Icon(color="green", icon="info-sign")
                 ).add_to(m)

    # 14) Añadir markers para hospitales dentro de cada buffer (con icono distintivo)
    # Aislamiento: pins rojos pequeños
    for _, h in hp_in_iso_wgs84.iterrows():
        popup_h = f"<b>{h.get(hosp_name_col,'Hospital')}</b><br>Distrito: {h.get('Distrito','-')}"
        folium.Marker(location=[h.geometry.y, h.geometry.x],
                      popup=folium.Popup(popup_h, max_width=300),
                      icon=folium.Icon(color="red", icon="plus-sign")
                     ).add_to(m)

    # Concentración: pins verdes pequeños
    for _, h in hp_in_con_wgs84.iterrows():
        popup_h = f"<b>{h.get(hosp_name_col,'Hospital')}</b><br>Distrito: {h.get('Distrito','-')}"
        folium.Marker(location=[h.geometry.y, h.geometry.x],
                      popup=folium.Popup(popup_h, max_width=300),
                      icon=folium.Icon(color="green", icon="plus-sign")
                     ).add_to(m)

    # 15) Capas y controles
    folium.LayerControl(collapsed=False).add_to(m)

    # 16) Guardar mapa
    if map_filename is None:
        map_filename = f"proximity_{region.lower()}.html"
    m.save(map_filename)
    print(f"Mapa de proximidad para {region_name} guardado en: {map_filename}")

    # 17) devolver resultados
    return {
        "map": m,
        "pop_reg_wgs84": pop_reg_wgs84,
        "buffers_wgs84": buffers_wgs84,
        "hp_in_iso_wgs84": hp_in_iso_wgs84,
        "hp_in_con_wgs84": hp_in_con_wgs84,
        "isolation": isolation,
        "concentration": concentration
    }

# ---------- Ejecutar para Lima y Loreto ----------
res_lima = make_proximity_map("LIMA", pop_gdf=pop_gdf, hp_gdf=hp_gdf, buffer_m=10000, map_filename="proximity_lima.html")
res_loreto = make_proximity_map("LORETO", pop_gdf=pop_gdf, hp_gdf=hp_gdf, buffer_m=10000, map_filename="proximity_loreto.html")
```
